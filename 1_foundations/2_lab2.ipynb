```python
request = "Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence. "
request += "Answer only with the question, no explanation."
messages = [{"role": "user", "content": request}]
```

```python
competitors = []
answers = []
messages = [{"role": "user", "content": question}]
```

```python
# The API we know well

model_name = "gpt-4o-mini"

response = openai.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Since I have API only for openAI I am going to use another model from their set 


model_name = "gpt-5-nano"

response = openai.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

## For the next cell, we will use Ollama

and runs models locally using high performance C++ code.

If you don't have Ollama, install it here by visiting https://ollama.com then pressing Download and following the instructions.

After it's installed, you should be able to visit here: http://localhost:11434 and see the message "Ollama is running"

You might need to restart Cursor (and maybe reboot). Then open a Terminal (control+\`) and run `ollama serve`

Useful Ollama commands (run these in the terminal, or with an exclamation mark in this notebook):

`ollama pull <model_name>` downloads a model locally  
`ollama ls` lists all the models you've downloaded  
`ollama rm <model_name>` deletes the specified model from your downloads

```python
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "llama3.2"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# So where are we?

print(competitors)
print(answers)
```

```python
# It's nice to know how to use "zip"
for competitor, answer in zip(competitors, answers):
    print(f"Competitor: {competitor}\n\n{answer}")
```

```python
# Let's bring this together - note the use of "enumerate"

together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
print(together)
```

```python
judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
# Judgement time!

openai = OpenAI()
response = openai.chat.completions.create(
    model="o3-mini",
    messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
```

```python
# OK let's turn this into results!

results_dict = json.loads(results)
ranks = results_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
```

```python
## Lets try to make a competition between two groups:
## One group is suggesting one single deep learning architecture for trajectory for autonomous vehicle 
## Teh other group is trying to find the best modular approach for trajectory planning 
## Each group will be consists of 3 members (same member) and a manager which combine their solution to a final solution which will be submitt to a judge (another LLM) and see how win 
```

Task: Provide a complete trajectory-planning solution for an autonomous vehicle. Inputs: (1) HD map data, (2) time-series histories of other road users, and (3) outputs from the perception stack (e.g., object detections, classifications, and semantic information).

```python

task = "Task: Provide a complete trajectory-planning solution for an autonomous vehicle. Inputs: (1) HD map data, (2) time-series histories of other road users, and (3) outputs from the perception stack (e.g., object detections, classifications, and semantic information)."

goup_number_1_task = task + " Your solution should consist of a single end-to-end network that takes perception inputs and map data as input and outputs a planned trajectory.No code only explanation what you suggest to do. The proposed approach should be competitive enough to win a benchmark competition."

group_number_2_task = task + " Your solution should follow a modular approach, providing separate networks for prediction, path planning, and velocity planning. These components should integrate into a multi-path planning and scoring system that selects the optimal trajectory. No code only explanation what you suggest to do.  And why it is good.The proposed approach should be competitive enough to win a benchmark competition"
```

```python
messages_group_1 = [{"role": "user", "content": goup_number_1_task}]
messages_group_2 = [{"role": "user", "content": group_number_2_task}]
```

```python
group_1_answers = []
group_2_answers = []
model_names = []
```

```python
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "llama3.2"

## Group 1 
response = ollama.chat.completions.create(model=model_name, messages=messages_group_1)
answer_g1 = response.choices[0].message.content


display(Markdown(answer_g1))

```

```python

## Group 2
response = ollama.chat.completions.create(model=model_name, messages=messages_group_2)
answer_g2 = response.choices[0].message.content
display(Markdown(answer_g2))
```

```python
model_names.append(model_name)
group_1_answers.append(answer_g1)
group_2_answers.append(answer_g2)
```

```python
model_name = "gpt-4o-mini"
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages_group_1,
)
answer_g1 = response.choices[0].message.content
display(Markdown(answer_g1))
```

```python
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages_group_2,
)
answer_g2 = response.choices[0].message.content
display(Markdown(answer_g2))
```

```python
model_names.append(model_name)
group_1_answers.append(answer_g1)
group_2_answers.append(answer_g2)
```

Now that you have 2 group member that answer the same question, we take a manager and he should come up with a final proposal which will be judge!!

```python
together_group_1 = ""
for index, answer in enumerate(group_1_answers):
    together_group_1 += f"# Response from emplyee {index+1}\n\n"
    together_group_1 += answer + "\n\n"

together_group_2 = ""
for index, answer in enumerate(group_2_answers):
    together_group_2 += f"# Response from emplyee {index+1}\n\n"
    together_group_2 += answer + "\n\n"
```

```python
group_1_message_to_manager = f"""You are the manager of two employees.  
Each employee has been given the following question:  

{messages_group_1}  

Your task is to combine their answers into a single proposal, which will be evaluated by the manager.  
The employees’ responses are as follows:  
{together_group_1}"""
```

```python
print(group_1_message_to_manager)
```

```python
group_2_message_to_manager = f"""You are the manager of two employees.  
Each employee has been given the following question:  

{messages_group_2}  

Your task is to combine their answers into a single proposal, which will be evaluated by the manager.  
The employees’ responses are as follows:  
{together_group_2}"""
```

```python
model_name = "gpt-5-nano"
openai = OpenAI()

messages_to_manager_1 = [{"role": "user", "content":group_1_message_to_manager}]
response = openai.chat.completions.create(
    model=model_name,
    messages=messages_to_manager_1,
)
answer_manager_group_1 = response.choices[0].message.content
display(Markdown(answer_manager_group_1))

```

```python
messages_to_manager_2 = [{"role": "user", "content":group_2_message_to_manager}]
response = openai.chat.completions.create(
    model=model_name,
    messages=messages_to_manager_2,
)
answer_manager_group_2 = response.choices[0].message.content
display(Markdown(answer_manager_group_2))
```

```python
both_proposals = f""" You are  given two proposal of two groups for solving the following task: {task}. Your goal is to decide which proposal is the best to go with because you are the boss who decide where we should put our effort. once you decide you should explain why you choose it. The proposal are the following: group numebr 1: {answer_manager_group_1}.

The proposal of group 2: {answer_manager_group_2}.
"""


```
